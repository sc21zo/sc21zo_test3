```python
import pandas as pd
from PIL import Image
import numpy as np
import os
import cv2
import random
import json
```

```python
df_label=pd.read_csv('../open-image/validation/label/detection.csv')# path of datasets detection
```

```python
Image_ids=df_label['ImageID'].unique()
```

```python
df_class=pd.read_csv('../open-image/validation/metadata/classes.csv',header=None)# path of datasets class
```

```python
with open('./dataset/val/validation/metadata/hierarchy.json','r') as f:
    t=json.load(f)
root_labels=t['Subcategory']
class_list=set()
for root_label in root_labels:
    if 'Subcategory' in root_label.keys():
        sub_labels=root_label['Subcategory']
        for sub_label in sub_labels:
            class_list.add(sub_label['LabelName'])
class_list=list(class_list)
```

```python
len(class_list)
```

```python
class_dict=dict()
for index,row in df_class.iterrows():
    class_dict[row[0]]=row[1]
```

```python
def is_exist(img_id,path):
    return os.path.exists(f'{os.path.join(path,img_id)}.jpg')
```

```python
image_path='../open-image/validation/data/'
```

```python
label_path='../open-image/validation/labels/'
```

```python
len(os.listdir(image_path))
```

```python
label_times=dict()
for l in class_list:
    label_times[l]=0
```

```python
for i,img_id in enumerate(Image_ids):
    img_labels=df_label[df_label['ImageID']==img_id]
    if len(img_labels)>0:
        if is_exist(img_id,image_path):
            count=0
            with open(os.path.join(label_path,img_id+'.txt'),'w') as f:
                
                for index,row in img_labels.iterrows():
                    LabelName=row['LabelName']
                    class_name=class_dict[LabelName]
                    if not class_name in class_list:
                        continue
                    if label_times[class_name]>=10:
                        continue
                    label_times[class_name]+=1
                    class_id=class_list.index(class_name)
                    XMin=float(row['XMin'])
                    XMax=float(row['XMax'])
                    YMin=float(row['YMin'])
                    YMax=float(row['YMax'])
                    XCentre=(XMin+XMax)/2.0
                    YCentre=(YMin+YMax)/2.0
                    box_w=abs(XMax-XMin)
                    box_h=abs(YMax-YMin)
                    box_info=' '.join([str(class_id),str(XCentre),str(YCentre),str(box_w),str(box_h)])
                    f.write(box_info+'\n')
                    count+=1
                if count==0:
                    os.remove(os.path.join(label_path,img_id+'.txt'))
                if i%100==0:
                    print(i)
```
